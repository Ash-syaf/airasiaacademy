## **Data Cleaning**

*Remove Unwanted Column*

```r
library(dplyr)
```

```r
newhousingpricetable=select(housingpricetable,c(transaction.year,Month,hoage,disMRT,no.cons,latitude,longitude,hopri))
```

```r
summary(newhousingpricetable)
```

Replacing NA categorical with mode

```r
# Create mode() function to calculate mode
mode <- function(x, na.rm = FALSE) {

  if(na.rm){ #if na.rm is TRUE, remove NA values from input x
    x = x[!is.na(x)]
  }

  val <- unique(x)
  return(val[which.max(tabulate(match(x, val)))])
}
```

```r
mode(newhousingpricetable$Month)
```

```r
newhousingpricetable$Month=ifelse(is.na(newhousingpricetable$Month),as.integer (mode(newhousingpricetable$Month)),newhousingpricetable$Month)
```

```r
newhousingpricetable$transaction.year=as.factor(newhousingpricetable$transaction.year)
```

```r
newhousingpricetable$Month=as.factor(newhousingpricetable$Month)
```

```r
summary(newhousingpricetable)
```
